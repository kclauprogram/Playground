# Question 2
Solve the initial value problem in Question 1 by numerical Euler method from t = 5 to t = 20. Take
the time step = 0.5.

i) Derive the approximate solution based on Euler’s method.

```python
import numpy as np

#get the increment of each step
def increment(t, y, step=0.5):
  slope = t/(9-t*t) * y
  return slope*step

#get the time seqence from 5 to 19.5
t = np.array(range(10, 40, 1))/2

#get the approximation of each step
appro = list([1])
for i in range(0,30):
  next = appro[i] + increment(t[i], appro[i])
  appro.append(next)

print(appro)
```

ii) Write down the value of y based on Euler’s method as well as the analytical solution in
Question 1 at t = 6, 8, 12, 18 and 20 and the corresponding relative error. (correct the answers
to 4 decimal places)

```python
def function(t):
  tmp = np.sqrt(t*t-9)
  return 4/tmp

time = [6,8,12,18,20]
timeslice = [2,6,14,26,30]

#get the approximated value
approximation = [appro[i] for i in timeslice]

#get the excat value
analytic = [function(i) for i in time]

#get their difference
diff = [round(approximation[i]-analytic[i], 4) for i in range(0,5)]

print(approximation)
print(analytic)
print(diff)
```

iii) Compare the result in Question 1 by plotting the analytical result in Question 1 and the
numerical result in Question 2.

We can observed that the approximation is undersetimating the analytic result all the time.

```python
import matplotlib.pyplot as plt

#get the x-axis from t=5 to t=20
x = np.array(range(10, 41, 1))/2

#get the analytic solution
a_result = [function(i) for i in x]

#plot the 2 result
plt.plot(x, appro, label="approximate solution")
plt.plot(x, a_result, label="analytic solution")
plt.legend()
plt.grid(True)
```

# Question 5 IV

Plot the responses in i, ii and iii and comment on the displacement in the cases I, II and III

The displacment of case I II III are decreasing as the frequency of force increasing

```python
import math
import numpy as np
import matplotlib.pyplot as plt

pi = math.pi

def func1(t):
  sub = 2*pi*math.sqrt(0.99)
  a = 5*math.cos(sub*t)
  b = 1/2/math.sqrt(0.99)*math.sin(sub*t)
  c = 5*math.cos(2*pi*t)
  return math.exp(-0.2*pi*t)*(a + b) - c


def func2(t):
  sub = 2*pi*math.sqrt(0.99)
  a = 4/22.9*math.cos(sub*t)
  b = 146/229/math.sqrt(0.99)*math.sin(sub*t)
  c1 = 4/22.9*math.cos(pi*t)
  c2 = 30/22.9*math.sin(pi*t)
  return math.exp(-0.2*pi*t)*(a - b) - c1 + c2

def func3(t):
  sub = 2*pi*math.sqrt(0.99)
  a = 1/557*math.cos(sub*t)
  b = 1201/5570/math.sqrt(0.99)*math.sin(sub*t)
  c1 = 1/557*math.cos(10*pi*t)
  c2 = 24/557*math.sin(10*pi*t)
  return math.exp(-0.2*pi*t)*(-a + b) - c1 - c2

x = np.array(range(0, 100, 1))/10

l1 = [func1(i) for i in x]

l2 = [func2(i) for i in x]

l3 = [func3(i) for i in x]

plt.plot(x, l1, label="line 1")
plt.plot(x, l2, label="line 2")
plt.plot(x, l3, label="line 3")
plt.legend()
plt.grid(True)
```
